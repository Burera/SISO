```python
import random
import numpy as np
import pdb
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
from math import *


M=2
Eb_N0_dB = 0
threshold = 0
num_of_sym = 0
BER_arr = []
symbol_cx = complex(0,0)
symbol_real = symbol_cx.real
symbol_imag = symbol_cx.imag
real_arr = []
imag_arr = []
symbol_tx_arr = []
symbol_rx_arr = []


for Eb_N0_dB in range (0,20):
    print('***********************************************************************************')
    print('Eb_N0_dB : ', Eb_N0_dB)
    Eb_N0_ratio = 10**(Eb_N0_dB/10)
    sigma = sqrt(1/(2*Eb_N0_ratio)) #Standard Deviation of Gaussian distribution
    error = 0
    num_of_sym = 0
    
    while(error < 100):
#1. x = 0 or not 1 create [0] [1]
        # 1. generating singal(symbol)
        symbol_tx = random.randrange(2) #0,1
        symbol_tx_arr.insert(0, symbol_tx)
            
#2. modulation 0 -> -1, 1 -> 1
        if (symbol_tx_arr[0] == 1):
            # symbol = cos(0) + sin(0)*1j  #1-> 1 + 0j
            symbol_real = cos(0)
            symbol_imag = sin(0)
        elif(symbol_tx_arr[0] == 0):
            # symbol = cos(pi) + sin(pi)*1j #0 -> -1 + 0j
            symbol_real = cos(pi)
            symbol_imag = 0
        # print("modulated symbol : ", symbol_tx)
        symbol = complex(symbol_real, symbol_imag)
        
        
        
#3. Multiply Rayleigh Channels and Add Noise
        # 3. generating noise in complex form
        AWGN_r = np.random.normal(0, sigma) #random value of a normal distribution that mean = 0, std = sigma
        AWGN = complex(AWGN_r, AWGN_r)
        
        # rayleigh_fading
        sigma_r = sqrt(0.5)
        x = np.random.normal(0, sigma_r)
        y = np.random.normal(0, sigma_r)
        h = complex(x, y)
        
        ## y = x*h + n
        x_h = symbol*h + AWGN  # yy = complex(symbol_real*x - symbol_imag*y + AWGN_r, symbol_real*y + symbol_imag*x + AWGN_i)

        ## y * h* = x * h * h* + n * h* = h* *(x*h + n)
        h_conj = h.conjugate()
        y_h_conj = x_h*h_conj  
        # y_h_conj = x_h.conjugate()*h_conj
        
        ## (y * h*)/\\h\\^2
        abs_square = pow(abs(h), 2)
        detection = y_h_conj/abs_square
        
     
        
#4. phase = atan(imag_tx/real_tx)
        #phase = atan2(detection.imag, detection.real)
        
#5. demodulation Depending on where the phase falls, right plane -> 1 left plane -> 0

        if(detection.real > 0):
            symbol_rx_arr.insert(0,1)
        elif(detection.real < 0):
            symbol_rx_arr.insert(0,0)
      
        
#6. information sinc first generated signal != demodulation signal -> error += 1
        if (symbol_rx_arr[0] != symbol_tx_arr[0]):
            error += 1
        num_of_sym += 1
     
        

    SER = error/num_of_sym   
    BER = SER/log2(M)
    BER_arr.append(BER)
    print(BER_arr)
   
plt.subplot(2,1,2)
plt.plot(BER_arr, 'bo-')
plt.xlim([0,22])
plt.yscale('log')
plt.xlabel('Eb/N0[dB]')
plt.ylabel('BER')
plt.title('BPSK Simulation_complex')
plt.grid(True)
plt.show()
   
```

```python
import random
import numpy as np
import pdb
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
from math import *

M=2
num_of_rx_atn = 2
SER_arr =[]
BER_arr = []
real_arr = []
imag_arr = []
symbol_tx_arr = []
symbol_rx_arr = []

for Eb_N0_dB in range (0,55):

    print('*********************************************************************************')
    print('Eb_N0_dB : ', Eb_N0_dB)
    Eb_N0_ratio = 10**(Eb_N0_dB/10)
    sigma = sqrt(1/(2*Eb_N0_ratio)) #Standard Deviation of Gaussian distribution
    # Eb_N0_dB += 1
    error = 0
    num_of_bit = 0

    while(error<100):

        # One. s1 s2 vector, 2X1 matrix
        s_1 = random.randrange(2)
        s_2 = random.randrange(2)
        symbol_tx_arr.insert(0, s_1)
        symbol_tx_arr.insert(1, s_2)
        # print("s1", s_1)
        # print("s2", s_2)

        #2. modulation
        if(s_1 == 0):
            symbol_real = cos(pi)
            symbol_imag = 0
        elif (s_1 == 1):
            symbol_real = cos(0)
            symbol_imag = sin(0)
            
        s_1_mod = complex(symbol_real, symbol_imag)
            
        if (s_2 == 0):
            symbol_real = cos(pi)
            symbol_imag = 0

        elif (s_2 == 1):
            symbol_real = cos(0)
            symbol_imag = sin(0)
            
        s_2_mod = complex(symbol_real, symbol_imag)

        symbol_tx_mod_arr = np.array([[s_1_mod, s_2_mod]])
        s = symbol_tx_mod_arr.T

        #3. H matrix , N(noise) vector
        #rayleigh_fading_channel
        sigma_r = sqrt(0.5)
        x11 = np.random.normal(0, sigma_r)
        y11 = np.random.normal(0, sigma_r)
        h11 = complex(x11,y11)
        x12 = np.random.normal(0, sigma_r)
        y12 = np.random.normal(0, sigma_r)
        h12 = complex(x12,y12)
        x21 = np.random.normal(0, sigma_r)
        y21 = np.random.normal(0, sigma_r)
        h21 = complex(x21,y21)
        x22 = np.random.normal(0, sigma_r)
        y22 = np.random.normal(0, sigma_r)
        h22 = complex(x22,y22)
        h_arr = np.array([[h11, h12], 
                          [h21, h22]])
        h_arr_norm = h_arr/sqrt(num_of_rx_atn)

        #AWGN
        AWGN_1 = np.random.normal(0, sigma)
        n_1 = complex(AWGN_1, AWGN_1)
        AWGN_2 = np.random.normal(0, sigma)
        n_2 = complex(AWGN_2, AWGN_2)
        AWGN_arr = np.array([[n_1, n_2]])
        AWGN = AWGN_arr.T
        #detection
        y = h_arr_norm @ s + AWGN
        h_inv = np.linalg.inv(h_arr_norm)
        h_inv_y = h_inv @ y
        s1_hat = h_inv_y[0,0]
        s2_hat = h_inv_y[1,0]
        
        #4. demodulation
        if (s1_hat.real > 0):
            symbol_rx_arr.insert(0,1)
        elif (s1_hat.real < 0):
            symbol_rx_arr.insert(0,0)
            
        if (s2_hat.real > 0):
            symbol_rx_arr.insert(1,1)
        elif (s2_hat.real < 0):
            symbol_rx_arr.insert(1,0)

        #5. information sinc
        
        if (symbol_tx_arr[0] != symbol_rx_arr[0]):
            error += 1
        if (symbol_tx_arr[1] != symbol_rx_arr[1]):
            error += 1
            
            
        num_of_bit += 2*log2(M)
        
        
        symbol_tx_arr = []
        symbol_rx_arr = []
        # pdb.set_trace()
        
        
    BER = error/num_of_bit    
    # SER = error/num_of_sym
    # SER_arr.append(SER)
    # BER = SER/log2(M)
    # norm_BER = BER*sqrt(num_of_rx_atn)
    BER_arr.append(BER)
    # pdb.set_trace()

    print(BER_arr)
    # pdb.set_trace()
    plt.plot(BER_arr, 'bo-')
plt.xlim([0, 50])
plt.yscale('log')
plt.xlabel('Eb/N0[dB]')
plt.ylabel('BER')
plt.title('MIMO2*2 Simulation_complex')
plt.grid(True)
plt.show()
```
